## Regex with Pandas

***Question:*** Load the .csv file 'sample.csv' and fill in the blanks. 'sample.csv' contains phone number and email address of members of a team.

```python
#import Pandas
import pandas as pd
# Load the sample.csv file

df = pd.read_csv('sample.csv')
```

```python
# This is how the data looks like
df
```

***Question1:*** Fill in the blanks so that we can extract only Argentina Phone numbers with ISD codes.
Note:
- I made up international code +22 for Argentina. 

```python
# Enter the regex to create the column 'Argentina_Number' and store only the number
# Result should look like below
argentina_numbers = df['Phone Number'].str.extractall(r"\+22-(?P<Argentine_Number>[\d\-]+)")
argentina_numbers
```

***Question2:*** Fill in the blanks so that we filter the Argentina numbers for only the valid Arg numbers.
Note:
- A valid argentinian number has 10 digits

```python
# Enter the regex to filter the argentiinian_numbers data frame for only valid numbers
# Result should look like below
argentina_numbers[argentina_numbers['Argentine_Number'].str.count(r'\d')==10]
```

***Question3:*** Fill in the blanks so that we extract the email username from the email ID column and store it into a separate column called 'Username'.
Note:
- For eg.: If the email ID is xyz@abc.com, create a column called 'Username' and store xyz in that

```python
# Enter the regex to extract the username from email ID
# Your output should look like the below
df['Email ID'].str.extractall(r'(?P<Username>.+)@')
```

***Question4*** Fill in the blanks so that we extract the email domain from the email ID column and store it into a separate column called 'Domain'.
Note:
- For eg.: If the email ID is xyz@abc.com, create a column called 'Domain' and store abc in that

```python
# Enter the regex to extract the domain from email ID
# Your output should look like the below
df['Email ID'].str.extractall(r'@(?P<Domain>\w+)\.')
```

***Question5*** Fill in the blanks so that we extract the primary non Argentinian numbers(without ISD code) with column name 'Primary_Number'.Non-Argentinian numbers have ISD code not equal to +22. For eg.: If a person has two numbers, Primary:+22-7879678433 and Secondary:+1-7584732222, the primary non-Argentinian number will be +1-7584732222.If a person has only one number then that should only be checked for being non-Argentinian or not. If a person has two numbers, Primary:+2-7846578911, Secondary:+1-2345657811, the primary non-Argentinian number will be +22-7846578911. If a person has two numbers, Primary:+1-3456712345, Secondary:+22-9876572345, the primary non-Argentinian number will be +1-3456712345.
Note:
- Observe that in the data, we have the numbers ordered, with primary numbers always before secondary numbers
- Do not bother if the number is not valid(count of digits not equal to 10)

```python
# Enter the regex to extract the non-Argentinian primary phone number from the Phone Number column without the ISD codes
# Your output should look like the below
df['Phone Number'].str.extract(r'(?<!\+91)-(?P<Primary_Number>\d+)')
```
