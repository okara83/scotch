```python
import sys
import pandas as pd
```

```python
df = pd.read_csv("drug_db.csv")
```

<b>sys.getsizeof( )</b> gives results in <b> bytes</b>

```python
sys.getsizeof(df)
```

<b>df.memory_usage( )</b> lists each column at $8\frac{\text{ bytes}}{\text{row}}$

```python
df.memory_usage()
```

 <b>df.memory_usage( ).sum( )</b> $\approx$ rows * cols * 8 bytes

```python
df.memory_usage().sum()
```

<b>df.memory_usage(deep=True) </b> lists each column's full memory usage

```python
df.memory_usage(deep=True)
```

<b>df.memory_usage(deep=True).sum()</b> gives results in bytes

```python
df.memory_usage(deep=True).sum()
```

<b>df.info() </b> prints dataframe info to stdout. Technically these are kibibytes (KiB), not kilobytes - as the docstring says, "Memory usage is shown in human-readable units (base-2 representation)." So to get bytes would multiply by 1024, e.g. 451.6 KiB = 462,438 bytes.

```python
df.info()
```

```python
df.info(memory_usage='deep')
```
