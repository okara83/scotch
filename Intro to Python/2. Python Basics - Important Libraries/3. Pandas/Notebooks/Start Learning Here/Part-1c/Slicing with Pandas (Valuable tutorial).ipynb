# NOTE
You might think that the code ref_df = df creates a fresh distinct copy of the df DataFrame object. However, using the = operator in the simple statement y = x does not create a copy of our DataFrame. Instead, y = x creates a new variable y that references the same object that x refers to. To state this another way, there is only one object (the DataFrame), and both x and y refer to it.
<br>
In contrast, the copy() method for a DataFrame creates a true copy of the DataFrame.
<b>
Letâ€™s look at what happens when we reassign the values within a subset of the DataFrame that references another DataFrame object:

# Slicing

```python
towns = df.loc[(df["embark_town"] == "Southampton"),  #this sets the condition to filter
               
               #line below selects only the header columns to keep after filter
               ["survived", "pclass", "sex", "age", "sibsp", "parch", "fare", 
                "embarked", "class", "alive", "alone"]]
                
```

```python
towns.head()
```

```python
towns = towns.sort_values("class",ascending=True)
towns = towns.reset_index(drop=True)
towns.head(50)
```

```python
towns.tail(50)
```

```python
towns_Second = towns.iloc[22:50,:]

#or: towns_Second = towns.iloc[[22,...,50],[3,...,]]
# where 3 = age,  4 = sibsp etc...
# Both of the above code snippets result in the following DataFrame:
```

```python
towns_Second.head(50)
```

```python
towns_Second = towns.iloc[22:50,3:]
towns_Second
```

Finally iloc[a,b] can also accept integer arrays as a and b, which is exactly why our second iloc example:

```python
towns_Second = towns.iloc[list(range(22,50)),list(range(3,11))]
```

```python
towns_Second
```

# See same as the previous example!
