# Python Lambda Function

# <u>Lambda</u> is an<b> anonymous function</b> in Python, that can accept any number of arguments, but can only have a single expression. 

###Lambda function is generally used in situations requiring an anonymous function for a short time period. Lambda functions can be used in either of the two ways:


# 1. Assigning lambda functions to a variable:

```python
mul = lambda a, b : a * b
print(mul(2, 5))
```

# 2. Wrapping lambda functions inside another function:

```python
def myWrapper(n):
    return lambda a : a * n
```

```python
mulFive = myWrapper(5)
```

```python
print(mulFive(2)) 
```
