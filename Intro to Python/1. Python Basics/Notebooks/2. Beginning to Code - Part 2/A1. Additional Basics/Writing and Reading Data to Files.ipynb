# Writing to Files

```python
path_and_name_of_file = 'file_to_write_in.txt'
file_obj = open(path_and_name_of_file, 'w')#write string to file opened in 'w' mode
string = 'write some string ending with a new in\n'
file_obj.write(string)
file_obj.close()
```

```python
#writing a string to a file opened in 'w' mode
file_obj = open('file_to_write_in.txt', 'w')
second_string = 'second line to write to file\n'
file_obj.write(second_string)
file_obj.close()
```

```python
!cat file_to_write_in.txt
```

Analogue of the previous command for <b>windows users</b>:<br>

```python
#writing a string to a file opened in 'a' mode
file_obj = open('file_to_write_in.txt', 'a')
second_string = 'third line to write to file\n'
file_obj.write(second_string)
file_obj.close()
```

```python
digits = range(1,11)
```

```python
digits
```

```python
#writing a list of lines to a file using the writelines function
file_obj = open('second_file_for_write_in.txt', 'w')
file_obj.writelines(digit + '\n' for digit in map(str, digits))
file_obj.close()
```

```python
#displaying the contents of a file
with open('second_file_for_write_in.txt', 'r') as file_obj:
    print(file_obj.read())
```
